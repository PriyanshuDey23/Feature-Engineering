# Ordinal numbering encoding or Label Encoding

## Ordinal Categorical Variables 

```python
import pandas as pd
import datetime
```

```python
# Create a variable with dates and from that , extract the weekdays
# I create a list of dates with 20 days difference from today
# and then transform it into a data frame

df_base = datetime.datetime.today()
df_date_list=[df_base - datetime.timedelta(days=x) for x in range (0,20)]
df=pd.DataFrame(df_date_list)
df.columns=['day']
df
```

```python
# Extract the weekday index

df['day_ordinal'] = df['day'].dt.weekday
#df['day_of_week']=df['weekday'].map({0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'})
df.head(7)
```

```python
# I have to create a dictionary for assigning the name of the week
weekday_map={6:'Sunday',0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday'}
df['day_of_week']=df.day_ordinal.map(weekday_map)
df.head(20)
              
```
